# Conceptual Exercise 1

## Question

We perform best subset, forward stepwise, and backward stepwise selection on a single data set. For each approach, we obtain $p + 1$ models, containing $0, 1, 2, \dots, p$ predictors. Explain your answers:

1. Which of the three models with $k$ predictors has the smallest *training* RSS?
2. Which of the three models with $k$ predictors has the smallest *test* RSS?
3. True or False:
    1. The predictors in the $k$-variable model identified by *forward* stepwise are a subset of the predictors in the $(k + 1)$-variable model identified by *forward* stepwise selection.
    2. The predictors in the $k$-variable model identified by *backward* stepwise are a subset of the predictors in the $(k + 1)$-variable model identified by *backward* stepwise selection.
    3. The predictors in the $k$-variable model identified by *backward* stepwise are a subset of the predictors in the $(k + 1)$-variable model identified by *forward* stepwise selection.
    4. The predictors in the $k$-variable model identified by *forward* stepwise are a subset of the predictors in the $(k + 1)$ variable model identified by *backward* stepwise selection.
    5. The predictors in the $k$-variable model identified by *best* subset are a subset of the predictors in the $(k + 1)$-variable model identified by *best* subset selection.

## Answer

1. For $k = 0$, all three models will be the null model $\mathcal{M}_0$, which predicts the sample mean for each observation. Similarly, for $k = p$, all three models will be the full model $\mathcal{M}_p$, which contains all $p$ predictors. In these two cases all three models will have the same training RSS. For $k = 1, 2, \dots, p - 1$, there may be situations where some of the models $\mathcal{M}_k$ contain the same predictors and therefore yield the same training RSS. In general, the model obtained through best subset selection will always have the minimum training RSS by construction, since in selecting $\mathcal{M}_k$ through best subset selection best is defined as having the smallest (training) RSS.

2. As noted in (1), when $k = 0$ and $k = p$ all three models are identical, and therefore will have the same test RSS. For $k = 1, 2, \dots, p - 1$, however, there are no guarantees as to which of the three models with $k$ predictors has the smallest test RSS. This is due to the fact that all three methods of subset selection use training RSS as the metric for determining the best model with $k$ predictors, but a model with low training error does not necessarily mean it will also have low test error.

3. Answers:
    
    1. True. In forward stepwise selection, the model $\mathcal{M}_{k + 1}$ is chosen by considering all $p - k$ models that add one additional predictor to the $k$-variable model $\mathcal{M}_k$. Then $\mathcal{M}_{k + 1}$ is the model out of these $p - k$ with the lowest (training) RSS.
    2. True. In backward stepwise selection, the model $\mathcal{M}_k$ is chosen by considering all $k + 1$ models that contain all but one of the predictors in the $(k + 1)$-variable model $\mathcal{M}_{k + 1}$. Then $\mathcal{M}_k$ is the model out of these $k + 1$ with the lowest (training) RSS.
    3. False. Consider a data set with $p = 3$ predictors. Suppose backward stepwise selection results in $\mathcal{M}^B_2$ containing $X_1$ and $X_2$, and $\mathcal{M}^B_1$ containing $X_1$. Further suppose forward stepwise selection results in $\mathcal{M}^F_1$ containing $X_3$, and $\mathcal{M}^F_2$ containing $X_2$ and $X_3$. In this situation, for $k = 1$, the predictors in the $k$-variable model identified by bacward stepwise ($X_1$) are not a subset of the predictors in the $(k + 1)$-variable model identified by forward stepwise selection ($X_2$ and $X_3$).
    4. False. Consider the same situation described above in (C). For $k = 1$, the predictors in the $k$-variable model identified by forward stepwise ($X_3$) are not a subset of the predictors in the $(k + 1)$-variable model identified by backward stepwise selection ($X_1$ and $X_2$).
    5. False. In best subset selection, model $\mathcal{M}_k$ is chosen by picking the model with the smallest (training) RSS out of all $\binom{p}{k}$ models with exactly $k$ predictors. Similarly, model $\mathcal{M}_{k + 1}$ is chosen by picking the model with the smallest (training) RSS out of all $\binom{p}{k + 1}$ models with exactly $k + 1$ predictors. Because each step involves choosing between all possible models containing a given number of predictors, there is no relationship between the predictors in $\mathcal{M}_k$ and those in $\mathcal{M}_{k + 1}$.

# Conceptual Exercise 2

## Question

For parts (1) through (3), indicate which of (A) through (D) is correct. Justify your answer.

1. The lasso, relative to least squares, is:
    1. More flexible and hence will give improved prediction accuracy when its increase in bias is less than its decrease in variance.
    2. More flexible and hence will give improved prediction accuracy when its increase in variance is less than its decrease in bias.
    3. Less flexible and hence will give improved prediction accuracy when its increase in bias is less than its decrease in variance.
    4. Less flexible and hence will give improved prediction accuracy when its increase in variance is less than its decrease in bias.
2. Repeat (1) for ridge regression relative to least squares.
3. Repeat (1) for non-linear methods relative to least squares.

## Answer

1. Relative to least squares, the lasso is less flexible because it imposes the additional condition $\sum_{j = 1}^p \lvert \beta_j \rvert \leq s$, where $s$ is a tuning parameter, on the coefficients $\beta_j$. This reduced flexibility increases the bias but lowers the variance. We can see this in the example for a very small value of $s$, which corresponds to a very large value of $\lambda$ in the usual description of the lasso penalty. In this case, the flexibility is reduced close to zero, so the resulting model is close to the null model which simply predicts the sample mean for each observation. The null model is extremely biased, as it only predicts a single value regardless of observation, but in that regard it has low variance and won't drastically change with small changes in the training sample. Therefore, (C) is correct. The lasso will give improved prediction accuracy when its increase in bias is less than its decrease in variance.

2. For the same reasons as discussed above in (1), ridge regression is less flexible than least squares. Ridge regression imposes the additional condition $\sum_{j = 1}^p \beta_j^2 \leq s$, where $s$ is a tuning paramter, on the coefficients $\beta_j$. Once again, this means that (C) is correct. Ridge regression will give improved prediction accuracy when its increase in bias is less than its decrease in variance.

3. As a result of their nonlinearity, non-linear methods are more flexible than least squares. The increased flexibility comes from fewer assumptions made about the true nature of the relationship between the predictors and the response, so non-linear methods have less bias compared to least squares. However, non-linear methods have more variance due to the increased flexibility. Thus, (B) is correct. Non-linear methods will give improved prediction accuracy when the increase in variance is less than the decrease in bias.

# Conceptual Exercise 3

## Question

Suppose we estimate the regression coefficients in a linear regression model by minimizing

\begin{equation}
    \sum_{i = 1}^n \left( y_i - \beta_0 - \sum_{j = 1}^p \beta_j x_{ij} \right)^2
    \quad \text{subject to} \quad
    \sum_{j = 1}^p \lvert \beta_j \rvert \leq s
\end{equation}

for a particular value of $s$. For parts (1) through (5) indicate which of (A) through (E) is correct. Justify your answer.

1. As we increase $s$ from 0, the training RSS will:
    1. Increase initially, and then eventually start decreasing in an inverted U shape.
    2. Decrease initially, and then eventually start increasing in a U shape.
    3. Steadily increase.
    4. Steadily decrease.
    5. Remain constant.
2. Repeat (1) for test RSS.
3. Repeat (1) for variance.
4. Repeat (1) for (squared) bias.
5. Repeat (1) for the irreducible error.

## Answer

Note that this optimization problem is that for the lasso. As we increase $s$ from 0, we go from the null model when $s = 0$, which predicts the sample mean for each observation, to the usual least squares regression model as $s$ approaches infinity.

1. As noted above, increasing $s$ from 0 takes us from the null model to the usual least squares model when $s$ is sufficiently large. The least squares model, by construction, is the model which results in the minimal training RSS. In addition, since increasing $s$ relaxes the constraint on the problem of minimizing the training RSS, the training RSS will steadily decrease (D).
2. The test RSS is a function of the variance plus the (squared) bias. The two quantities change at different rates as $s$ increases from 0. Initially the increase in variance will be small relative to the decrease in bias, until the point where variance begins to increase substantially relative to the decreasing bias. Thus, the test RSS will decrease initially, and then eventually start increasing in a U shape (B).
3. Increasing $s$ gives more room for the coefficients $\beta_j$ to vary, so the variance will steadily increase (C).
4. Another way of saying that increasing $s$ gives more room for the coefficients $\beta_j$ to vary is to say that it reduces the contraints on them, and therefore reduces the assumptions made regarding the model. Thus, the (squared) bias will steadily decrease (D).
5. The irreducible error comes from the inherent error term in the function for the true relationship between the predictors and the response. Since it is unrelated to the choice of model, it will always remain constant (E).

# Conceptual Exercise 5

## Question

It is well-known that ridge regression tends to give similar coefficient values to correlated variables, whereas the lass may give quite different coefficient values to correlated variables. We will now explore this property in a very simple setting.

Suppose that $n = 2$, $p = 2$, $x_{11} = x_{12}$, and $x_{21} = x_{22}$. Furthermore, suppose that $y_1 + y_2 = 0$, $x_{11} + x_{21} = 0$, and $x_{12} + x_{22} = 0$, so that the estimate for the intercept in a least squares, ridge regression, or lasso model is zero: $\hat{\beta}_0 = 0$.

1. Write out the ridge regression optimization problem in this setting.
2. Argue that in this setting, the ridge coefficient estimates satsify $\hat{\beta}_1 = \hat{\beta}_2$.
3. Write out the lasso optimization problem in this setting.
4. Argue that in this setting, the lasso coefficients $\hat{\beta}_1$ and $\hat{\beta}_2$ are not unique -- in other words, there are many possible solutions to the optimization problem in (3). Describe these solutions.

## Answer

1. In this situation, with ridge regression we are trying to minimize the value of 

\begin{align}
    f(\beta_1, \beta_2)  
    &= \sum_{i = 1}^2 \left(y_i - \sum_{j = 1}^2 \beta_j x_{ij} \right)^2 + \lambda \sum_{j = 1}^2 \beta_j^2 \\
    &= (y_1 - \beta_1 x_{11} - \beta_2 x_{12})^2 + (y_2 - \beta_1 x_{21} - \beta_2 x_{22})^2 + \lambda(\beta_1^2 + \beta_2^2).
\end{align}

2. In order to minimize $f(\beta_1, \beta_2)$, we take partial derivatives with respect to each variable and set each equal to zero.

\begin{align}
    \frac{\partial f}{\partial \beta_1} &= -2x_{11}(y_1 - \beta_1 x_{11} - \beta_2 x_{12}) - 2x_{21}(y_2 - \beta_1 x_{21} - \beta_2 x_{22}) + 2\lambda \beta_1 \\
    \frac{\partial f}{\partial \beta_2} &= -2x_{12}(y_1 - \beta_1 x_{11} - \beta_2 x_{12}) - 2x_{22}(y_2 - \beta_1 x_{21} - \beta_2 x_{22}) + 2\lambda \beta_2
\end{align}

2. *(cont)* Note that since $x_{11} = x_{12}$ and $x_{21} = x_{22}$, we can subtract the two partial derivatives above to obtain the equation $2\lambda\beta_1 - 2\lambda\beta_2 = 0$. In other words, $f(\beta_1, \beta_2)$ is minimized when $\beta_1 = \beta_2$. This gives us ridge coefficient estimates that satisfy $\hat{\beta}_1 = \hat{\beta}_2$.

3. We'll use the alternate phrasing of optimization problem. In this situation, with the lasso we are trying to minimize the value of $g(\beta_1, \beta_2)$ subject to the constraint $\lvert \beta_1 \rvert + \lvert \beta_2 \rvert \leq s$ for some value of $s \geq 0$, where

\begin{align}
    g(\beta_1, \beta_2)  
    &= \sum_{i = 1}^2 \left(y_i - \sum_{j = 1}^2 \beta_j x_{ij} \right)^2\\
    &= (y_1 - \beta_1 x_{11} - \beta_2 x_{12})^2 + (y_2 - \beta_1 x_{21} - \beta_2 x_{22})^2.
\end{align}

3. *(cont)* Geometrically, we are trying to minimize $g(\beta_1, \beta_2)$ within the diamond formed by the lines $\beta_2 = s - \beta_1$, $\beta_2 = \beta_1 - s$, $\beta_2 = -s - \beta_1$, and $\beta_2 = \beta_1 + s$ in the $(\beta_1, \beta_2)$-plane. See the left portion of Figure 6.7 of the textbook for a visual.

4. To find the absolute minimum value of a continuous function $g$ on a closed, bounded set $D$, we first find the values of $g$ at the critical points of $g$ inside $D$. As we did in Part 2 above, this involves taking the partial derivatives with respect to each variable, setting them equal to zero, and then solving for $\beta_1$ and $\beta_2$.

\begin{align}
    \frac{\partial g}{\partial \beta_1} &= -2x_{11}(y_1 - \beta_1 x_{11} - \beta_2 x_{12}) - 2x_{21}(y_2 - \beta_1 x_{21} - \beta_2 x_{22})\\
    \frac{\partial g}{\partial \beta_2} &= -2x_{12}(y_1 - \beta_1 x_{11} - \beta_2 x_{12}) - 2x_{22}(y_2 - \beta_1 x_{21} - \beta_2 x_{22})
\end{align}

4. *(cont)* Since $x_{11} = x_{12}$ and $x_{21} = x_{22}$, these two partial derivatives are in equal, which means that instead of there being a unique critical point, there is an entire line of critical points.

\begin{equation}
    0 = \frac{\partial g}{\partial \beta_1}
    = (x_{11}y_1 + x_{21}y_2) - (x_{11}^2 + x_{21}^2)\beta_1 - (x_{11}^2 + x_{21}^2)\beta_2
\end{equation}

4. *(cont)* Following the convention we used earlier for writing lines in the $(\beta_1, \beta_2)$-plane, the line of critical points is given by the following equation.

\begin{equation}
    \beta_2 = \frac{x_{11}y_1 + x_{21}y_2}{x_{11}^2 + x_{21}^2} - \beta_1
\end{equation}

4. *(cont)* On this line of critical points, the value of $g$ is constant.

\begin{align}
    g(\beta_1, \beta_2) &= \left( y_1 - (\beta_1 + \beta_2)x_{11} \right)^2 + \left( y_2 - (\beta_1 + \beta_2)x_{21} \right)^2 \\
    &= \left( y_1 - \left( \frac{x_{11}y_1 + x_{21}y_2}{x_{11}^2 + x_{21}^2} \right)x_{11} \right)^2 + \left( y_2 - \left( \frac{x_{11}y_1 + x_{21}y_2}{x_{11}^2 + x_{21}^2} \right)x_{21} \right)^2
\end{align}

4. *(cont)* Note that this line is parallel to the sides $\beta_2 = s - \beta_1$ and $\beta_2 = -s - \beta_1$ in the diamond of constraints. The second step is to analyze $g$ on the boundary. On the first boundary piece, $\beta_2 = s - \beta_1$, we see that $g(\beta_1, \beta_2) = (y_1 - sx_{11})^2 + (y_2 - sx_{21})^2$ is constant. Similarly, on the third boundary piece, $\beta_2 = -s - \beta_1$, $g(\beta_1, \beta_2) = (y_1 + sx_{11})^2 + (y_2 + sx_{21})^2$ is constant. On the remaining two boundary pieces, $g$ is not constant. Plugging in the equation for the second boundary piece, $\beta_2 = \beta_1 - s$, taking the derivative with respect to $\beta_1$ and then solving when the derivative is equal to zero, we get an extreme value occuring at

\begin{equation}
    \beta_1 = \frac{x_{11}y_1 + x_{21}y_2}{2(x_{11}^2 + x_{21}^2)} + \frac{s}{2}, \quad
    \beta_2 = \frac{x_{11}y_1 + x_{21}y_2}{2(x_{11}^2 + x_{21}^2)} - \frac{s}{2}.
\end{equation}

4. *(cont)* Following the same process for the fourth boundary piece $\beta_2 = \beta_1 + s$, we see that the extreme value of $g$ on that boundary piece occurs at

\begin{equation}
    \beta_1 = \frac{x_{11}y_1 + x_{21}y_2}{2(x_{11}^2 + x_{21}^2)} - \frac{s}{2}, \quad
    \beta_2 = \frac{x_{11}y_1 + x_{21}y_2}{2(x_{11}^2 + x_{21}^2)} + \frac{s}{2}.
\end{equation}

4. *(cont)* Both of these points on the boundary give the same value for $g(\beta_1, \beta_2)$, which is equal to the value of $g$ on the line of critical points. This analysis gives three scenarios, and in each scenario there are not unique values for $\beta_1$ and $\beta_2$ that minimize $g$. The first scenario is when $g$ is minimized on the line of critical points. In addition to the entire line of values which minimize $g$ in this case, $g$ is also minimized at the two points on the boundary pieces $\beta_2 = \beta_1 \pm s$ indicated above. The second and third scenarios are when $g$ is minimized on the boundary pieces $\beta_2 = s - \beta_1$ and $\beta_2 = -s - \beta_1$, respectively.

# Conceptual Exercise 6

## Question

We will now explore optimization problems for ridge regression and the lasso in the simple case where $n = p$, $X$ is the identity matrix (i.e. diagonal with 1's on the diagonal and 0's in all off-diagonal elements), and that we are performing regression without an intercept. In this setting, ridge regression amounts to finding $\beta_1, \dots, \beta_p$ such that

\begin{equation}
    \sum_{j = 1}^p (y_j - \beta_j)^2 + \lambda \sum_{j = 1}^p \beta_j^2
\end{equation}

is minimized, and the lasso amounts to finding the coefficients such that

\begin{equation}
    \sum_{j = 1}^p (y_j - \beta_j)^2 + \lambda \sum_{j = 1}^p \lvert \beta_j \rvert
\end{equation}

is minimized.

1. Consider the optimization problem for ridge regression with $p = 1$. For some choice of $y_1$ and $\lambda > 0$, plot the optimization problem as a function of $\beta_1$. Your plot should confirm that the optimization problem is solved by 

\begin{equation}
    \hat{\beta}_1^R = \frac{y_1}{1 + \lambda}.
\end{equation}

2. Consider the optimization problem for the lasso with $p = 1$. For some choice of $y_1$ and $\lambda > 0$, plot the optimization problem as a function of $\beta_1$. Your plot should confirm that the optimization problem is solved by 

\begin{equation}
    \hat{\beta}_1^L = 
    \begin{cases}
        y_1 - \lambda / 2 & \text{if } y_1 > \lambda / 2;\\
        y_1 + \lambda / 2 & \text{if } y_1 < -\lambda / 2;\\
        0 & \text{if } \lvert y_1 \rvert \leq \lambda / 2.
    \end{cases}
\end{equation}

## Answer

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
lambdas = (np.logspace(-2, 2, num = 4)).reshape(2, 2)
beta = np.linspace(-50, 50)
rng = np.random.default_rng(312)
y = rng.random()*50
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (12, 12), constrained_layout = True)
fig_title = r"Plots of ridge regression error vs $\hat{\beta}^R_1$ for " + "\n" + r"$y = {0:.2f}$ and various values of $\lambda$".format(y)
fig.suptitle(fig_title, fontsize = 16)
for i in range(lambdas.shape[0]):
    for j in range(lambdas.shape[1]):
        axes[i, j].plot(beta, (y - beta)**2 + lambdas[i, j]*beta**2)
        beta_min = y/(1 + lambdas[i, j])
        min_error = (y - beta_min)**2 + lambdas[i, j]*beta_min**2
        axes[i, j].plot([beta_min], [min_error], "o")
        annotation = "Minimum at\n" + r"$\hat{\beta}^R_1 = $" + "{0:.2f}".format(beta_min)
        axes[i, j].annotate(annotation, xy = (beta_min, min_error), xytext = (0, 50), textcoords = "offset points",
                           arrowprops=dict(facecolor = "black"), horizontalalignment = "center")
        axes[i, j].set_title(r"$\lambda = {0:.2f}$".format(lambdas[i, j]))
        axes[i, j].set_xlabel(r"$\hat{\beta}^R_1$")
        axes[i, j].set_ylabel("Ridge regression error")
```

```python
def one_d_ridge_min(y, l):
    if y > l/2:
        return y - l/2
    elif y < -l/2:
        return y + l/2
    else:
        return 0
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (12, 12), constrained_layout = True)
fig_title = r"Plots of lasso regression error vs $\hat{\beta}^L_1$ for " + "\n" + r"$y = {0:.2f}$ and various values of $\lambda$".format(y)
fig.suptitle(fig_title, fontsize = 16)
for i in range(lambdas.shape[0]):
    for j in range(lambdas.shape[1]):
        axes[i, j].plot(beta, (y - beta)**2 + lambdas[i, j]*np.abs(beta))
        beta_min = one_d_ridge_min(y, lambdas[i, j])
        min_error = (y - beta_min)**2 + lambdas[i, j]*abs(beta_min)
        axes[i, j].plot([beta_min], [min_error], "o")
        annotation = "Minimum at\n" + r"$\hat{\beta}^L_1 = $" + "{0:.2f}".format(beta_min)
        axes[i, j].annotate(annotation, xy = (beta_min, min_error), xytext = (0, 50), textcoords = "offset points",
                           arrowprops=dict(facecolor = "black"), horizontalalignment = "center")
        axes[i, j].set_title(r"$\lambda = {0:.2f}$".format(lambdas[i, j]))
        axes[i, j].set_xlabel(r"$\hat{\beta}^L_1$")
        axes[i, j].set_ylabel("Lasso regression error")
```
